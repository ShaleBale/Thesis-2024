```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from itertools import cycle
import numpy as np
```

```python
file_path = r'C:\Users\shlbs\OneDrive - The University of Texas at Austin\Desktop\SRS_Thesis_ShareFolder\_DATA\_AOIs\_CORNUDAS\_FIELD SPEC\Cornudas_Processed_FieldSpec\CND031\CND031_Average.csv'
```

```python
 df = pd.read_csv(file_path, delimiter=',')
```

```python
df.head()
```

```python
plt.figure(figsize=(14, 6))
plt.scatter(df['Wavelength (nm)'], df['Average Reflectance'], color='blue', linewidth=0.5)
plt.title('Scatter plot of Wavelength vs. Average Reflectance')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Average Reflectance')
plt.grid(True)
plt.show()
```

```python
derivative = (df['Average Reflectance'].diff() / df['Wavelength (nm)'].diff()).fillna(0)

# Plot the derivative
plt.figure(figsize=(8, 6))
plt.plot(df['Wavelength (nm)'], derivative, color='green')
plt.title('Derivative of Average Reflectance with respect to Wavelength')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Derivative')
plt.grid(True)
plt.show()
```

```python
# Create figure and axes
fig, ax1 = plt.subplots(figsize=(8, 6))

# Plot original data on ax1
ax1.plot(df['Wavelength (nm)'], df['Average Reflectance'], color='blue', label='Original Data', linewidth=1.0)

# Create a secondary y-axis for the derivative
ax2 = ax1.twinx()
ax2.plot(df['Wavelength (nm)'], derivative, color='green', label='Derivative')

# Set labels for each y-axis
ax1.set_ylabel('Average Reflectance')
ax2.set_ylabel('Derivative')

# Set common x-axis label and title
ax1.set_xlabel('Wavelength (nm)')
ax1.set_title('Wavelength vs. Average Reflectance with Derivative')

# Add legend for both sets of data
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

plt.grid(True)
plt.show()
```
