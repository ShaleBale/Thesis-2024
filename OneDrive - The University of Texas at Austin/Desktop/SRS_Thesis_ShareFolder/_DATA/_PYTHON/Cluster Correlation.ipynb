```python
columns_for_correlation = ['Ig_Cluster_2', 'Sed_Cluster_DER_4']
```

```python
for cluster_column in columns_for_correlation:
    print(f"Correlation analysis for DataFrame based on '{cluster_column}':")
    current_df = dfs[cluster_column]  # Get the corresponding DataFrame from the dictionary
    for column in current_df.columns:
        if column != cluster_column:  # Exclude correlation with itself
            correlation, p_value = spearmanr(current_df[cluster_column], current_df[column], nan_policy='omit')
            if not pd.isna(correlation) and p_value < 0.05:  # Check for significant correlation
                print(f"   '{cluster_column}' vs '{column}': Spearman correlation = {correlation:.2f}, p-value = {p_value:.4f}")
            else:
                print(f"   '{cluster_column}' vs '{column}': Not significant (p-value = {p_value:.4f})")
```

```python
# Initialize a dictionary to store significant correlations for each cluster
significant_correlations = {}

# Iterate through each cluster
for cluster_column in columns_for_correlation:
    print(f"Correlation analysis for DataFrame based on '{cluster_column}':")
    current_df = dfs[cluster_column]  # Get the corresponding DataFrame from the dictionary
    significant_correlations[cluster_column] = {}
    
    # Iterate through each column in the current DataFrame
    for column in current_df.columns:
        if column != cluster_column:  # Exclude correlation with itself
            correlation, p_value = spearmanr(current_df[cluster_column], current_df[column], nan_policy='omit')
            
            # Check for significant correlation
            if not pd.isna(correlation) and p_value < 0.05:
                # Store significant correlations in the dictionary
                significant_correlations[cluster_column][column] = {'correlation': correlation, 'p-value': p_value}

# Print significant correlations for each cluster
for cluster, correlations in significant_correlations.items():
    print(f"Significant correlations for cluster '{cluster}':")
    for column, values in correlations.items():
        print(f"   '{cluster}' vs '{column}': Spearman correlation = {values['correlation']:.2f}, p-value = {values['p-value']:.4f}")
```

```python
# Define the columns for comparison, excluding 'Ig_Cluster_2'
columns_for_comparison = [col for col in df_Ig_Cluster_2.columns if col != 'Ig_Cluster_2']

for column in columns_for_comparison:
    print(f"Comparing 'Ig_Cluster_2' against '{column}':")
    if df_Ig_Cluster_2[column].dtype == 'category':
        # Perform chi-square test of independence for two categorical variables
        contingency_table = pd.crosstab(df_Ig_Cluster_2['Ig_Cluster_2'], df_Ig_Cluster_2[column])
        chi2, p_value, _, _ = chi2_contingency(contingency_table)
        if p_value < 0.05:  # Check for significance
            print(f"   'Ig_Cluster_2' vs '{column}': Significant difference (p-value = {p_value:.4f})")
        else:
            print(f"   'Ig_Cluster_2' vs '{column}': Not significant (p-value = {p_value:.4f})")
    else:
        # Perform appropriate statistical test for numerical vs. categorical comparison
        unique_categories = df_Ig_Cluster_2['Ig_Cluster_2'].cat.categories
        all_values_identical = all(df_Ig_Cluster_2[df_Ig_Cluster_2['Ig_Cluster_2'] == cat][column].nunique() == 1 for cat in unique_categories)
        if all_values_identical:
            print(f"   'Ig_Cluster_2' vs '{column}': All numbers are identical within at least one group. Unable to perform Kruskal-Wallis test.")
        elif len(unique_categories) > 2:
            # Perform ANOVA
            stat, p_value = f_oneway(*[df_Ig_Cluster_2[df_Ig_Cluster_2['Ig_Cluster_2'] == cat][column] for cat in unique_categories])
            if p_value < 0.05:  # Check for significance
                print(f"   'Ig_Cluster_2' vs '{column}': Significant difference (p-value = {p_value:.4f})")
            else:
                print(f"   'Ig_Cluster_2' vs '{column}': Not significant (p-value = {p_value:.4f})")
        else:
            # Perform Kruskal-Wallis test
            stat, p_value = kruskal(*[df_Ig_Cluster_2[df_Ig_Cluster_2['Ig_Cluster_2'] == cat][column] for cat in unique_categories])
            if p_value < 0.05:  # Check for significance
                print(f"   'Ig_Cluster_2' vs '{column}': Significant difference (p-value = {p_value:.4f})")
            else:
                print(f"   'Ig_Cluster_2' vs '{column}': Not significant (p-value = {p_value:.4f})")
```

```python
# Initialize an empty list to store significant comparisons
significant_comparisons = []

for column in columns_for_comparison:
    if df_Ig_Cluster_2[column].dtype == 'category':
        # Perform chi-square test of independence for two categorical variables
        contingency_table = pd.crosstab(df_Ig_Cluster_2['Ig_Cluster_2'], df_Ig_Cluster_2[column])
        chi2, p_value, _, _ = chi2_contingency(contingency_table)
        if p_value < 0.05:  # Check for significance
            significant_comparisons.append((f"Ig_Cluster_2 vs {column}", "Chi-square", chi2, p_value))
    else:
        # Perform appropriate statistical test for numerical vs. categorical comparison
        unique_categories = df_Ig_Cluster_2['Ig_Cluster_2'].cat.categories
        all_values_identical = all(df_Ig_Cluster_2[df_Ig_Cluster_2['Ig_Cluster_2'] == cat][column].nunique() == 1 for cat in unique_categories)
        if not all_values_identical:
            if len(unique_categories) > 2:
                # Perform ANOVA
                stat, p_value = f_oneway(*[df_Ig_Cluster_2[df_Ig_Cluster_2['Ig_Cluster_2'] == cat][column] for cat in unique_categories])
                test_name = "ANOVA"
            else:
                # Perform Kruskal-Wallis test
                stat, p_value = kruskal(*[df_Ig_Cluster_2[df_Ig_Cluster_2['Ig_Cluster_2'] == cat][column] for cat in unique_categories])
                test_name = "Kruskal-Wallis"
            
            if p_value < 0.05:  # Check for significance
                significant_comparisons.append((f"Ig_Cluster_2 vs {column}", test_name, stat, p_value))

# Create a DataFrame from the list of significant comparisons
significant_results_df = pd.DataFrame(significant_comparisons, columns=['Comparison', 'Test', 'Test Statistic', 'P-value'])

# Display the DataFrame
print(significant_results_df)
```
