```python
import os
import pandas as pd
import matplotlib.pyplot as plt

all_wavelengths = []
all_reflectance = []
folder_names = []

# Iterate through subfolders
for foldername in os.listdir(base_folder):
    folder_path = os.path.join(base_folder, foldername)
    
    if os.path.isdir(folder_path):
        excel_file_name = foldername + '_Average.csv'  # Create the CSV file name based on the folder name
        excel_file_path = os.path.join(folder_path, excel_file_name)
        
        # Check if the CSV file exists
        if os.path.exists(excel_file_path):
            # Read the CSV file into a pandas DataFrame
            df = pd.read_csv(excel_file_path)
            
            # Extract data from the DataFrame
            wavelengths = df['Wavelength (nm)']  # Use the actual column name
            reflectance = df['Average Reflectance']  # Use the actual column name
            
            # Accumulate data and folder names
            all_wavelengths.append(wavelengths)
            all_reflectance.append(reflectance)
            folder_names.append(foldername)

# Create a scatter plot for all data with unique colors
plt.figure()
for i in range(len(all_wavelengths)):
    color_map = plt.get_cmap('viridis')
    color = color_map(i / len(all_wavelengths))  # Get a unique color for each dataset
    plt.plot(all_wavelengths[i], all_reflectance[i], label=folder_names[i], color=color, linewidth=1.0)

    # Calculate midpoints for the lines
    mid_x = all_wavelengths[i].mean()
    mid_y = all_reflectance[i].mean()

    # Annotate lines with sample numbers at the midpoints if the Series are not empty
    if not all_wavelengths[i].empty and not all_reflectance[i].empty:
        plt.text(mid_x, mid_y, folder_names[i], fontsize=8, color=color, ha='center', va='bottom')

plt.xlabel('Wavelength (nm)', fontsize=18)
plt.ylabel('Average Reflectance', fontsize=18)

legend = plt.legend(loc='upper left', title='Sample Number', ncol=2, bbox_to_anchor=(1.05, 1))
legend.get_title().set_fontsize(20)
legend.get_frame().set_alpha(None)
legend.get_frame().set_linewidth(0.0)
```
