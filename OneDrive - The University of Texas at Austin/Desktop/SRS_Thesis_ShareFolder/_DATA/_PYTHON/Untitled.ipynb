```python
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
station_column = 'Station ID'
wavelength_column = 'Wavelength (nm)'
reflectance_column = 'Average Reflectance'

# Scatter plot for each station
for station_id, group in df.groupby(station_column):
    plt.scatter(group[wavelength_column], group[reflectance_column], label=station_id)

# Customize plot labels and title
plt.xlabel('Wavelength (nm)')
plt.ylabel('Average Reflectance')
plt.title('Victorio Peak Formation Data')
plt.legend()  # Show legend with station IDs

# Show the plot
plt.show()
```

```python
mean_by_wavelength = df.groupby(wavelength_column)[reflectance_column].mean().reset_index()
median_by_wavelength = df.groupby(wavelength_column)[reflectance_column].median().reset_index()

# Display the mean DataFrame
print("Mean DataFrame:")
print(mean_by_wavelength)

# Display the median DataFrame
print("\nMedian DataFrame:")
print(median_by_wavelength)
```

```python
plt.figure(figsize=(10, 6))
for station_id, group in df.groupby(station_column):
    plt.scatter(group[wavelength_column], group[reflectance_column], label=station_id, linewidth = 0.005)

# Plot mean values
mean_by_wavelength = df.groupby(wavelength_column)[reflectance_column].mean().reset_index()
plt.plot(mean_by_wavelength[wavelength_column], mean_by_wavelength[reflectance_column], marker='o', linestyle='-', label='Mean', color='black', linewidth=2.0)

# Plot median values
median_by_wavelength = df.groupby(wavelength_column)[reflectance_column].median().reset_index()
plt.plot(median_by_wavelength[wavelength_column], median_by_wavelength[reflectance_column], marker='x', linestyle='--', label='Median', color='red', linewidth=2.0)

# Customize plot labels and title
plt.xlabel('Wavelength (nm)', fontsize=18)
plt.ylabel('Average Reflectance', fontsize=18)
plt.title('Scatter Plot with Mean and Median Values Across Samples', fontsize=18)
plt.legend(loc='upper left', title='Station ID', bbox_to_anchor=(1.05, 1), fontsize=12, frameon=False, facecolor=(0.9, 0.9, 0.9, 0.7))

# Show the plot
plt.tight_layout()
plt.show()
```

```python
variance_by_wavelength = df.groupby(wavelength_column)[reflectance_column].std()

# Plot variance
plt.figure(figsize=(10, 6))
plt.plot(variance_by_wavelength.index, variance_by_wavelength.values, marker='o', linestyle='-', color='green', linewidth=2.0)

# Customize plot labels and title
plt.xlabel('Wavelength (nm)', fontsize=18)
plt.ylabel('Reflectance Variance', fontsize=18)
plt.title('Reflectance Variance Across Wavelengths', fontsize=18)

# Show the plot
plt.grid(True)
plt.show()
```
