```python
grouped = df.groupby('Field_ID')

# Initialize a list to store time series arrays
time_series_data = []

# Iterate over groups and extract time series data
for group_name, group_data in grouped:
    # Extract Wavelength and Average Reflectance columns
    wavelengths = group_data['Wavelength (nm)'].values
    reflectances = group_data['Average Reflectance'].values
    
    # Create a time series array
    time_series_array = np.column_stack((wavelengths, reflectances))
    
    # Append to the list
    time_series_data.append(time_series_array)

# Convert the list of time series arrays to a numpy array
time_series_dataset = np.array(time_series_data)
```

```python
# Define the range of clusters
range_n_clusters = range(2, 11)

# Initialize lists to store the within-cluster sum of squares (WCSS) for each clustering method
wcss_kmeans = []

# Calculate WCSS for KMeans
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(time_series_2d)
    wcss_kmeans.append(kmeans.inertia_)

# Plot the elbow method for KMeans
plt.plot(range_n_clusters, wcss_kmeans, label='KMeans', marker='o')

plt.title('Elbow Method for KMeans Clustering Algorithm')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.legend()
plt.show()
```

```python
range_n_clusters = [2, 3, 4, 5,]

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(time_series_2d) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(time_series_2d)

    silhouette_avg = silhouette_score(time_series_2d, cluster_labels)
    print("For n_clusters =", n_clusters, "The average silhouette_score is :", silhouette_avg)

    sample_silhouette_values = silhouette_samples(time_series_2d, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values, facecolor=color, edgecolor=color, alpha=0.7)
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10  

    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")
    ax1.set_yticks([])
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    ax1.set_title("Silhouette plot for {} clusters".format(n_clusters))
    ax1.set_xlabel("Silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # 2nd Plot showing the actual clusters formed
    ax2.scatter(time_series_2d[:, 0], time_series_2d[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=cluster_labels, cmap='viridis', edgecolor="k")

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    ax2.scatter(centers[:, 0], centers[:, 1], marker="o", c="white", alpha=1, s=200, edgecolor="k")

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

    ax2.set_title("Cluster visualization for {} clusters".format(n_clusters))
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(
        "Silhouette analysis for KMeans clustering on sample data with {} clusters".format(n_clusters),
        fontsize=14,
        fontweight="bold",
    )

plt.show()
```

```python
neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(time_series_2d)
distances, indices = nbrs.kneighbors(time_series_2d)
distances = np.sort(distances, axis=0)
distances = distances[:,1]

# Plot the knee/elbow graph
plt.plot(distances)
plt.title('Knee/Elbow Method for DBSCAN')
plt.xlabel('Clusters')
plt.ylabel('Epsilon')
plt.show()

# Define epsilon_values and wcss_dbscan based on the computed distances
epsilon_values = np.arange(len(distances))
wcss_dbscan = distances

# Now you can use epsilon_values and wcss_dbscan in the Plotly code
fig = go.Figure()

# Add a scatter plot for the WCSS values
fig.add_trace(go.Scatter(x=epsilon_values, y=wcss_dbscan, mode='lines+markers', name='DBSCAN WCSS'))

# Add a vertical line for the knee/elbow point
knee_index = 2  # You need to set the index of the knee/elbow point
fig.add_shape(type="line", x0=epsilon_values[knee_index], y0=min(wcss_dbscan), 
              x1=epsilon_values[knee_index], y1=max(wcss_dbscan),
              line=dict(color="red", width=2, dash="dash"))

# Add text annotation for the knee/elbow point
fig.add_annotation(x=epsilon_values[knee_index], y=min(wcss_dbscan), text="Knee/Elbow Point", 
                    showarrow=True, arrowhead=1)

# Customize layout
fig.update_layout(title='Elbow Method for DBSCAN',
                  xaxis_title='Clusters',
                  yaxis_title='Epsilon',
                  hovermode='x unified')

# Show the interactive plot
fig.show()
```

```python
# Preprocess the data (mean-variance normalization)
X_train = TimeSeriesScalerMeanVariance().fit_transform(time_series_2d)

# Define the number of clusters (k)
n_clusters = 3

# Initialize and fit the TimeSeriesKMeans model
km = TimeSeriesKMeans(n_clusters=n_clusters, metric="euclidean", verbose=True, random_state=42)
km.fit(X_train)

# Get cluster assignments for the data
cluster_labels = km.predict(X_train)

# Print cluster centers
print("Cluster centers:")
for i, center in enumerate(km.cluster_centers_):
    print("Cluster", i, "center:", center)

# Print cluster assignments for all samples
print("Cluster assignments for all samples:")
for i, label in enumerate(cluster_labels):
    print("Sample", i, "belongs to cluster", label)
```

```python
# Plot original time series colored by cluster
plt.figure(figsize=(10, 6))
for i in range(n_clusters):
    plt.subplot(n_clusters, 1, i + 1)
    for j in range(len(cluster_labels)):
        if cluster_labels[j] == i:
            plt.plot(time_series_2d[j].ravel(), color='C'+str(i), alpha=0.5)
    plt.plot(km.cluster_centers_[i], color='k', linewidth=2)  # Plot cluster center without ravel()
    plt.title(f'Cluster {i+1}')
plt.tight_layout()
plt.show()
```
